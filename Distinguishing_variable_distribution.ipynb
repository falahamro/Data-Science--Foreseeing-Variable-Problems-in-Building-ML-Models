We will visualize the variable distributions

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the boston house prices dataset from scikit-learn 

from sklearn.datasets import load_boston
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, 
                      columns=boston_dataset.feature_names)
```

```python
# visulaize the variable distribution with histograms: 

boston.hist(bins=30, figsize=(12,12), density=True)
plt.show()
```

Most of the numerical variables in the dataset are skewed.
